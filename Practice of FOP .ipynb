```python
# use a nested for loop to print a right-angled triangle of stars.
for i in range(1, 6):
    for j in range(i):
        print("*", end="")
    print()
```

```python
def hollow_sqaure(n):
    for i in range(n):
        for j in range(n):
            if i == 0 or i == n - 1 or j == 0 or j == n-1:
                print("*", end=" ")
            else:
                print(" ", end=" ")
        print()
hollow_sqaure(5)
```

```python
 # Find the largest number in the list [4, 9, 2, 10, 7] using a for loop.
 my_list = [4, 9, 2, 10, 7]
 largest = my_list[0]
 for num in my_list:
    if num > largest:
        largest = num
        print(my_list)
print(f"The largest number is {largest}")
```

```python
 # Count how many even and odd numbers are present in [1, 2, 3, 4, 5, 6].
 my_list = [1, 2, 3, 4, 5, 6]
 even_count = 0
 odd_count = 0
 for num in my_list:
    if num % 2 == 0:
        even_count += 1
    else:
        odd_count += 1
        print(f"Even numbers: {even_count}")
        print(f"Odd numbers: {odd_count}")
```
