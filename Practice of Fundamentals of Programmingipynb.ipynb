```python
# 1. Create a dictionary and print it.
# Question 1 : create a dictionary that stores the name and age of a person.
person={"name": "rahul", "age" : 20}
print(person)
```

```python
# 2. Access a value from a dictionary.
# Question 2 : given the dictionary {"name": "rahul", "age": 20}, print the age.
person={"name":"rahul", "age": 20}
print(person["age"])
```

```python
# 3. Add a new key-value pair to a dictionary.
# Question 3 : Add a key "city" with value "new york" to the dictionary person.
person = {"name":"rahul","age" : 20}
person["city"]="new york"
print(person)
```

```python
## Updating an existing value in a dictionary.
# Question 4 : change the age of "alice" to 28 in the given dictionary.
person={"name": "rahul","age": 20}
person["age"]=25
print(person)
```

```python
# Delete a key from a dictionary.
# Question 5 : remove  the "age" from the dictionary.
person={"name": "rahul","age": 20}
del person["age"]
print(person)
```

```python
# Get the length of a dictionary.
# Question 6 : find how many items are in the dictionary.
person={"name": "rahul","age": 20}
print(len(person))
```

```python
# Copy a dictionary.
# Question 7 : create a copy of the dictionary person.
person={"name": "rahul","age": 20}
copy_person= person.copy()
print(copy_person)
```

```python
# Use get () method to access value safely.
# Question 8 :  Access the "city" key using get(), provide default "unknown" if not present.
person={"name": "rahul","age": 20}
city=person.get("city","unknown")
print(city)
```
